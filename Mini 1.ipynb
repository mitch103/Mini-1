```python
import pandas as pd
import numpy as np

import plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)




School_Fund = pd.read_csv('/Users/markmitchell/Downloads/userssharedsdfschoolimprovement2010grants.csv')
School_Fund
```

```python
School_Fund.columns = ['School', 'City', 'State', 'District', 'Amount', 'Model', 'Location']
School_Fund
```

```python
School_Fund['Amount'] = School_Fund['Amount'].str.replace('$','')
```

```python
School_Fund['Amount'] = School_Fund.Amount.astype(float)
```

```python
School_Fund.describe()
```

```python
df = School_Fund[School_Fund['Amount']>School_Fund['Amount'].mean()]
df
```

```python
pd.options.display.max_colwidth = 500
```

```python
School_Fund.Location.str.split("\n").head()
```

```python
df["Latitude"]  = df.Location.str.split("\n").apply(pd.Series)[2].str.replace("(", "").str.replace(")", "").str.split(",").apply(pd.Series)[0].astype(float)
df["Longitude"] = df.Location.str.split("\n").apply(pd.Series)[2].str.replace("(", "").str.replace(")", "").str.split(",").apply(pd.Series)[1].astype(float)
```

<Font Size = 5> **Funding across the United States**

```python
School_Fund.columns = ['School', 'City', 'State', 'District', 'Amount', 'Model', 'Location']
School_Fund
```

```python
iplot(
    go.Figure(
        data = [
            go.Bar(x= School_Fund['State'], y = School_Fund['Amount']),
        ],
        
        layout = go.Layout(
        title = '2010-2011 School Funding Awards',
        yaxis = dict(title = 'Amount'),
        xaxis = dict(title  = 'States')
        )
        ))
```

From this bar graph, we can see the max funding of each state as well as the average. The four states that stand out the most are Loiusana, Maine, Rhode Island, and Texas because there was no funding in these states during this time period. Why would that be? Were taxes high enough to the point where they didn't need Funding from the government or do they receive private funding?

```python
iplot(
    go.Figure(
        data = [
            go.Bar(x= School_Fund['Model'], y = School_Fund['Amount']),
        ],
        
        layout = go.Layout(
        title = '2010-2011 School Funding Awards',
        yaxis = dict(title = 'Amount'),
        xaxis = dict(title  = 'Model')
        )
        ))
```

From this bar graph, we can see that the model that has the highest average funding is the restart model. It's just short of the transformation model at just under a max of 1 million dollars. We can also see that of the 4 models, closing schools received the least amount of funding.
